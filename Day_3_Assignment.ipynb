```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.shape
```

```python
fmri.info()
```

```python
fmri.describe()
```

```python
sns.relplot(x = 'timepoint', y= 'signal', kind = 'scatter', data = fmri, legend = 'full')
```

```python
sns.relplot(x='timepoint', y = 'signal', data = fmri, hue= 'region', style='event')
```

```python
sns.relplot(x = 'timepoint', y= 'signal', kind = 'line', data = fmri)
```

```python
sns.catplot(x ='event', y='signal', data = fmri, kind = 'boxen')
```

```python
sns.catplot(x ='event', y='signal', data = fmri, kind = 'box')
```

```python
sns.catplot(x ='signal', y='event', data = fmri, kind = 'box', orient= 'horizontal')
```

```python
sns.catplot(x ='event', y='signal', data = fmri, kind = 'swarm' )
```

```python
sns.catplot(x ='event', y='signal', data =fmri , kind = 'violin')
```
